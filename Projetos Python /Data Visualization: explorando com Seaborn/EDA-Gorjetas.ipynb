# Tratamento de dados 

```python
import pandas as pd 
```

```python
pd.read_csv('dados/tips.csv')
```

```python
dados = pd.read_csv('dados/tips.csv')
```

```python
dados.head()
```

```python
renomear = {
    'total_bill' : 'valor_da_conta',
    'tip' : 'gorjeta',
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana',
    'time' : 'hora_do_dia',
    'size' : 'total_de_pessoas'
}
```

```python
gorjetas = dados.rename(columns = renomear)
```

```python
gorjetas.head(11)
```

```python
gorjetas.sobremesa.unique()
```

```python
sim_nao = {
    'No' : 'Não',
    'Yes' : 'Sim'
}
```

```python
gorjetas.sobremesa = gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas.head(1)
```

```python
gorjetas.dia_da_semana.unique()
```

```python
dias = {
    'Sun' : 'Domingo',
    'Sat' : 'Sábado',
    'Thur' : 'Quinta',
    'Fri' : 'Sexta'
}
```

```python
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias)
```

```python
gorjetas.head(1)
```

```python
gorjetas.hora_do_dia.unique()
```

```python
hora = {
    'Dinner' : 'Jantar',
    'Lunch' : 'Almoço'
}
```

```python
gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
```

```python
gorjetas.head()
```

## Criando o campo porcentagem

```python
gorjetas.head(1)
```

```python
gorjetas['porcentagem'] = gorjetas['gorjeta'] / gorjetas['valor_da_conta']
```

```python
gorjetas.porcentagem = gorjetas.porcentagem.round(2)
```

```python
gorjetas.head(1)
```

```python
porcentagem_conta = sns.scatterplot(x='valor_da_conta', y='porcentagem', data=gorjetas)
```

### O resultado é bastante interessante, pois visualmente parece que o valor da conta não é proporcional ao valor da gorjeta, e sim que diminui. Ou seja, nas contas mais altas, os valores das gorjetas foram maiores do que as mais baixas, porém não são proporcionais, apesar das impressões do primeiro gráfico feito anteriormente.

# Teste de Hipótese

```python
from scipy.stats import ranksums
```

```python
gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python

sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
ranksums(sobremesa, sem_sobremesa)
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print('O valor do p-value é {}'.format(r.pvalue))
```
